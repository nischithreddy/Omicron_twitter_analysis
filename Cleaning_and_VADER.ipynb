**Perform text cleaning on tweets**
The text cleaning will be limited as VADER reads text with certain emojis or punctuations and assigns score based on these factors as well.

```python


nltk.download('punkt')
nltk.download('stopwords')

STOP_WORDS = stopwords.words()

remove_punc = "\"#$%&'()+,-./:;<=>?@[\]^_`{|}~"     # we kept ! and * because they are used by people to express sentiments

# dfx = dfx[['created_at', 'id', 'text']]

# Read the Omicron tweets that are clubbed as weekly files
interesting_files = glob.glob("/content/drive/MyDrive/IDS 506 Healthcare Analytics/506 Project/Omicron_Tweets/*.csv") 
#print(len(interesting_files))
number_of_files = len(interesting_files)

for week in range(1, number_of_files+1):    # number_of_files
  path = "/content/drive/MyDrive/IDS 506 Healthcare Analytics/506 Project/Omicron_Tweets/Omi_Week" + str(week) + ".csv"
  df = pd.read_csv(path, engine= 'python')
  #df = df[['created_at', 'id', 'omicron_tweets', 'user_id']]

  EMOJI_PATTERN = re.compile("["
                            #u"\U0001F600-\U0001F64F"  # emoticons
                            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                            u"\U0001F680-\U0001F6FF"  # transport & map symbols
                            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                            u"\U00002702-\U000027B0"
                            u"\U000024C2-\U0001F251"
                            "]+", flags=re.UNICODE)


  def cleaning(text):
      '''
      Convert to lowercase.
      Rremove URL links, special characters and punctuation.
      Tokenize and remove stop words.
      '''
      #text = text.lower()
      text = re.sub('https?://\S+|www\.\S+', '', text)
      
      text = re.sub('@[A-Za-z0-9]+', '', text)
      text = re.sub('<.*?>+', '', text)   # replacing HTML text with blank or whitespace
      
      text = re.sub('\n', ' ', text)
      #text_try = re.sub(r'[^\w\s]', '', text_try)
      text = re.sub('[%s]' % re.escape(remove_punc), '', text)
      #text = re.sub('[%s]' % re.escape(string.punctuation), '', text)     
      # removes the puncutation characters. re.escape is used to escape the special characters that form puncuation characters. 
      #For eg, \" considers quote as character because of backslash
      
      text = re.sub('[’“”…]', '', text)
      
    # text = text.encode("ascii", "ignore")     # consider only english alphanumeric characters
    # text = text.decode()
      
      text = EMOJI_PATTERN.sub(r'', text)
      return text
    
      
  '''  
      # removing the stop-words
      text_tokens = word_tokenize(text)
      tokens_without_sw = [
        word for word in text_tokens if not word in STOP_WORDS]
      filtered_sentence = (" ").join(tokens_without_sw)
      text = filtered_sentence
  '''
      
      


  #if __name__ == "__main__":
  #    max_rows = 1000  # 'None' to read whole file
      
  df['omicron_tweets'] = df['omicron_tweets'].astype(str)
  #df_reduced.head() 

  #dt = df_reduced['name_plus_desp'].apply(cleaning)
  df['clean_text'] = df['omicron_tweets'].apply(cleaning)

  analyzer = SentimentIntensityAnalyzer()

  # dfx['Sentiment_Score'] = dfx['text'].apply(lambda x: analyser.polarity_scores(x))
  df['compound'] = [analyzer.polarity_scores(x)['compound'] for x in df['clean_text']]
  df['neg'] = [analyzer.polarity_scores(x)['neg'] for x in df['clean_text']]
  df['neu'] = [analyzer.polarity_scores(x)['neu'] for x in df['clean_text']]
  df['pos'] = [analyzer.polarity_scores(x)['pos'] for x in df['clean_text']]

  path2 = "/content/drive/MyDrive/IDS 506 Healthcare Analytics/506 Project/Vader_OP"

  #os.makedirs(path, exist_ok=False)
  isExist = os.path.exists(path2)
  if not isExist:
  
    # Create a new directory because it does not exist 
    os.makedirs(path2)
    print("The new directory is created!")
  #dest_path = "/content/drive/MyDrive/IDS 506 Healthcare Analytics/506 Project/Clean_tweets/Clean_Week_" + rng + ".csv"
  dest_path2 = "/content/drive/MyDrive/IDS 506 Healthcare Analytics/506 Project/Vader_OP/V_Sentiment_Week" + str(week) + ".csv"
  #df.to_csv(dest_path)
  df.to_csv(dest_path2, index=False)

df.head(900)

```
